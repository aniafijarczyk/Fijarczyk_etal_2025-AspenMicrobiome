```python
dtax = pd.read_csv("../../2022_MicrobiomeAspenGreenhouse/01_data_prep/dataPrep_01_overview_16S_TAX.tsv", sep="\t", header=0)
print(dtax.shape)
dtax = dtax.fillna("unid")
#dtax = dtax.drop(columns=['Species'])
dtax['Domain'] = dtax['Domain'].str.replace('d__','')
dtax['Phylum'] = dtax['Phylum'].str.replace('p__','')
dtax['Class'] = dtax['Class'].str.replace('c__','')
dtax['Order'] = dtax['Order'].str.replace('o__','')
dtax['Family'] = dtax['Family'].str.replace('f__','')
dtax['Genus'] = dtax['Genus'].str.replace('g__','')
dtax['Species'] = dtax['Species'].str.replace('s__','')
dtax = dtax.replace("uncultured","unid")
dtax = dtax.reset_index().rename(columns={'index':'ASV'})
dtax['genus_taxon'] = dtax.apply(lambda x: x['Domain']+'_'+x['Phylum']+'_'+x['Class']+'_'+x['Order']+'_'+x['Family']+'_'+x['Genus'], axis=1)
dtax.head()
```

```python
dtax['Genus'].drop_duplicates().sort_values().values
```

# Groupby up to genus, add consecutive number & ungroup

```python
unique_genera = dtax.groupby(['Domain','Phylum','Class','Order','Family','Genus','genus_taxon']).agg(n = ('ASV','count')).reset_index()
print(unique_genera.shape)
unique_genera.head()
```

# Adding consecutive numbers to the unique genera

```python
i=1
new_genus = []
for ind, row in unique_genera.iterrows():
    genus = row['Genus']
    if genus == 'unid':
        new_gen = 'unid'+'_'+str(i)
    else:
        new_gen = genus
    new_genus.append(new_gen)
    i+=1
new_genus
unique_genera['Genus_'] = new_genus
unique_genera.head()
```

# Naming unidentified

```python
full_genus = []
for ind, row in unique_genera.iterrows():
    genus = row['Genus']
    if genus == 'unid':
        previous = [row['Domain'],row['Phylum'], row['Class'], row['Order'], row['Family']]
        previous_nonan = [ele for ele in previous if ele != 'unid']
        if previous_nonan:
            descriptor = previous_nonan[-1]
            full_gen = row['Genus_'] + "_" + descriptor
            #full_gen = row['Genus_']
        else:
            full_gen = row['Genus_']
    else:
        full_gen = row['Genus_']
    full_genus.append(full_gen)
unique_genera['taxon'] = full_genus
unique_genera.head()
```

# Putting back genus name to ASV

```python
mtax = pd.merge(dtax[['ASV','genus_taxon']],unique_genera,on='genus_taxon', how='left')
print(mtax.shape)
mtax = mtax[['ASV','taxon','Domain','Phylum','Class','Order','Family','Genus','genus_taxon']].reset_index(drop=True).set_index('ASV')
mtax.head()
```
