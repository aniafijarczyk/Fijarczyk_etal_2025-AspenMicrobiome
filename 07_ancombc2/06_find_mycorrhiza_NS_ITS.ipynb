```python
import pandas as pd
import numpy as np
```

# DATA

```python
demf = pd.read_csv('../../2022_MicrobiomeQuebecMexico/DATA/ITS/Full/Peuplier_REPRISE_ITS_oct2022.ASV_table_norarefaction_dnNA.guilds.txt', sep='\t', header=0)
demf = demf.loc[demf['Confidence Ranking'].isin(["Probable","Highly Probable"]), ['Taxon','Guild']].drop_duplicates().reset_index(drop=True)
demf['Guild'] = demf['Guild'].fillna("unknown")
print(demf.shape)
demf.head()
demf['MF'] = demf['Guild'].apply(lambda x: "MF" if "mycorrhiz" in x else "non-MF")
#demf = demf[demf["MF"] == "MF"].reset_index(drop=True)
demf['Guild'].drop_duplicates().values
demf.head()
```

# ANCOMBC2 Results

```python
dres = pd.read_csv("ancombc_04_run_ITS_results_random_effect.tsv", sep="\t", header=0)
dres.head()
dres['taxon_last'] = dres['taxon_org'].apply(lambda x: x.split('_')[-1])
print(dres.shape)
dres.head()
dres['taxon_last'].drop_duplicates().values
```

```python
dm = pd.merge(dres, demf, left_on = "taxon_last", right_on = "Taxon", how = "left")
print(dm.shape)
dm["MF"] = dm["MF"].fillna("Unknown")
dm["MF"] = dm["MF"].replace("non-MF","Other")
dm["MF"] = dm["MF"].replace("MF","Mycorrhizal")
dm.head()
```

```python
dm.to_csv("ancombc_09_ITS_EMF.tsv", sep="\t", header=True, index=False)
```
